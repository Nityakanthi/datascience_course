**Problem 10**

Categorizes vehicles into 'Light', 'Medium', and 'Heavy' based on 'weight_kg'. Use the following criteria: 'Light' (< 1500 kg), 'Medium' (1500-2000 kg), 'Heavy' (> 2000 kg). Use the `apply` method to create a new column 'weight_category' based on this custom function.

Analyze the results to determine the relationship between vehicle weight/weight_category and fuel efficiency.


### Solution

+ Converting weight of vehicle from pounds to kgs
+ categorizing the weight of vehicle accorifing to criteria
+ calling the function using apply
+ print the vehicle

```python
vehicle['weight_kg']=vehicle['weight']*0.453592
```

```python
def categorize(weight):
    value = weight['weight']*0.453592
    if value < 1500:
        v_weight = 'Light'
    elif value > 2000:
        v_weight = 'Heavy'
    else:
        v_weight = 'Medium'
    return v_weight
```

```python
vehicle['weight_category'] = vehicle.apply(categorize, axis = 1)
```
